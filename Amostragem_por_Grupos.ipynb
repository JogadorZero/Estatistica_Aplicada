```python
import pandas as pd
import random
import numpy as np

dataset = pd.read_csv('census.csv')
```

#Amostragem por grupos

#####Na amostragem por grupos precisamos inicialmente definir a quantidade de grupos que vamos ter na anÃ¡lise.

```python
len(dataset)
```

```python
len(dataset)/10
```

```python
grupos = []
id_grupo = 0
contagem = 0
for _ in dataset.iterrows():
  grupos.append(id_grupo)
  contagem += 1
  if contagem > 3256:
    contagem = 0
    id_grupo += 1
```
